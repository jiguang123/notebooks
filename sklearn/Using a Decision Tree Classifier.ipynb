# Using a Decision Tree Classifier

```python
import numpy as np

%matplotlib inline
```

```python
from sklearn.datasets import load_iris

iris = load_iris()

x, y = iris.data, iris.target
```

```python
from sklearn.cross_validation import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.4, random_state=4)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(max_depth=4).fit(xtrain, ytrain)
```

## Visualizing the Tree

```python
from io import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus


def treeviz(fitted_tree): 
    dot_data = StringIO()  
    export_graphviz(fitted_tree, out_file=dot_data, 
                    feature_names=iris.feature_names,  
                    class_names=iris.target_names, rounded=True)  
    graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
    return Image(graph.create_png())
```

```python
treeviz(tree)
```
